# Prueba simple de RAG - Mundiales

Notebook minimo para probar la funcion de consumo definida en `src/rag/RAG.py`.

```python
from pathlib import Path
import sys

# Resolver raiz del proyecto para poder importar src.rag.RAG
BASE_DIR = Path.cwd().resolve()
for candidate in [BASE_DIR, *BASE_DIR.parents]:
    if (candidate / "src").exists() and (candidate / "data").exists():
        BASE_DIR = candidate
        break

if str(BASE_DIR) not in sys.path:
    sys.path.append(str(BASE_DIR))

from src.rag.RAG import consumir_rag_mundiales

print("BASE_DIR:", BASE_DIR)
```

```python
# Prueba rapida de una sola pregunta
resultado = consumir_rag_mundiales(
    "Quien gano el mundial de 2010?",
    llm_model="llama3.2:3b",
    top_k=4,
)

print("Respuesta:\n", resultado["respuesta"])
print("\nChunks usados:", resultado["chunks"])
print("Fuentes:", resultado["fuentes"][:2])
```

```python
# Pruebas estilo lista (similar a tus guardrails)
preguntas_prueba = [
    "Quien gano el mundial de 1930?",
    "Quien fue campeon en 1998?",
    "En que pais se jugo el mundial de 2014?",
    "Cual fue el resultado de la final de 2010?",
]

for pregunta in preguntas_prueba:
    out = consumir_rag_mundiales(pregunta, llm_model="llama3.2:3b", top_k=4)
    print(f"Pregunta: {pregunta}")
    print(f"Respuesta: {out['respuesta']}")
    print(f"Chunks: {out['chunks']}")
    print("-" * 70)
```

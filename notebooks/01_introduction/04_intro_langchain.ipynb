## Notebook 4: Introducción a LangChain con Ollama

**LangChain** es un framework diseñado para simplificar la creación de aplicaciones que utilizan Modelos de Lenguaje de Gran Escala (LLMs).

1. **Modelos de Chat (ChatModels):** Envolturas unificadas para interactuar con LLMs (usaremos `ChatOllama`).
2. **Plantillas de Prompts (PromptTemplates):** Formas dinámicas de construir instrucciones.
3. **Parsers de Salida (Output Parsers):** Herramientas para formatear las respuestas del modelo.
4. **Cadenas y LCEL (LangChain Expression Language):** La sintaxis para encadenar los componentes usando el operador pipe (`|`).

## 3. Modelos de Chat (ChatModels)

Inicializaremos el modelo utilizando la clase `ChatOllama` provista por LangChain, que se conectará automáticamente a nuestro servidor local.

```python
from langchain_ollama import ChatOllama

# Inicializar el modelo local de Ollama (solo si el servidor está corriendo)
model = ChatOllama(model="qwen2.5:1.5b", temperature=0.7)
print("Modelo ChatOllama cargado y listo.")
```

---  

## 4. Plantillas de Prompts (Prompt Templates)

Creamos una plantilla parametrizable para estructurar las instrucciones del sistema y del usuario de manera formal.

```python
from langchain_core.prompts import ChatPromptTemplate

# Definir la estructura del prompt
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "Eres un chef experto. Responde brevemente con una receta clásica del país indicado."),
    ("user", "Dame una idea de plato para {pais}.")
])

# Formatear el prompt de prueba
messages = prompt_template.format_messages(pais="México")
print("Mensajes del prompt formateado:")
for msg in messages:
    print(f"[{msg.type}]: {msg.content}")
```

---  

## 5. Parsers de Salida (Output Parsers)

Utilizaremos `StrOutputParser` para extraer automáticamente el texto de la respuesta estructurada del modelo.

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

---  

## 6. Cadenas con LCEL (LangChain Expression Language)

Usamos el operador `|` para crear nuestro pipeline secuencial: `prompt | model | parser`. Observa cómo la sintaxis de LangChain es idéntica independientemente de si usas un modelo en la nube o local con Ollama.

```python
# Encadenar componentes
chain = prompt_template | model | parser

# Invocar la cadena
resultado = chain.invoke({"pais": "Italia"})
print("--- RESPUESTA DE LA CADENA CON OLLAMA ---")
print(resultado)
```

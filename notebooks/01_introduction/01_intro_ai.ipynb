## 1. El Panorama de la Inteligencia Artificial

La Inteligencia Artificial es un campo amplio que ha evolucionado significativamente a lo largo de las décadas. A menudo se representa como círculos concéntricos:

*   **Inteligencia Artificial (IA):** Programas con capacidad de aprender y razonar como humanos (sistemas expertos, lógica difusa, algoritmos de búsqueda).
*   **Machine Learning (ML):** Algoritmos cuyas prestaciones mejoran a medida que se les expone a más datos a lo largo del tiempo.
*   **Deep Learning (DL):** Subcampo del ML que utiliza redes neuronales artificiales de múltiples capas (redes profundas) para aprender representaciones complejas.
*   **Generative AI (GenAI):** Modelos capaces de generar nuevos contenidos (texto, imágenes, código, audio) a partir de patrones aprendidos en datos de entrenamiento.

| Concepto | Enfoque Principal | Ejemplo de Uso |
| :--- | :--- | :--- |
| **IA Tradicional** | Reglas codificadas a mano | Motor de ajedrez basado en reglas |
| **Machine Learning** | Patrones estadísticos | Predicción del precio de viviendas |
| **Deep Learning** | Representaciones jerárquicas | Reconocimiento facial en imágenes |
| **GenAI / LLMs** | Creación de contenido nuevo | Chatbots de soporte al cliente |

## 2. Tipos de Aprendizaje Automático (ML)

Los algoritmos de Machine Learning se clasifican principalmente en tres categorías basadas en cómo aprenden de los datos:

### A. Aprendizaje Supervisado (Supervised Learning)
El modelo entrena con datos **etiquetados** (cada entrada viene con su respuesta correcta).
*   **Regresión:** Predecir un valor continuo (ej. precio de una casa).
*   **Clasificación:** Predecir una clase o categoría discreta (ej. si un correo es Spam o No Spam).

### B. Aprendizaje No Supervisado (Unsupervised Learning)
El modelo entrena con datos **no etiquetados**. Busca patrones o estructuras ocultas.
*   **Clustering (Agrupamiento):** Agrupar clientes por comportamiento de compra similar.
*   **Reducción de Dimensionalidad:** Simplificar datos complejos sin perder información relevante.

### C. Aprendizaje por Refuerzo (Reinforcement Learning)
Un agente aprende a tomar decisiones interactuando con un entorno para maximizar una recompensa.
*   **Ejemplo:** Robots aprendiendo a caminar, agentes jugando videojuegos.

## 3. Del Dato a la Predicción: Conceptos Clave

Para que una máquina aprenda, debemos representar el mundo en términos matemáticos:
*   **Características (Features - $X$):** Las variables de entrada que describen el problema (ej. tamaño de casa, número de habitaciones).
*   **Etiqueta (Label - $y$):** La variable objetivo que queremos predecir (ej. precio de la casa).
*   **Pesos (Weights - $w$):** Parámetros que determinan la importancia de cada característica.
*   **Sesgo (Bias - $b$):** Un valor constante que permite ajustar la predicción hacia arriba o hacia abajo de forma independiente de las características.

La fórmula más básica de predicción lineal es:
$$\hat{y} = w_1 x_1 + w_2 x_2 + ... + w_n x_n + b$$

O en notación matricial/vectorial:
$$\hat{y} = XW^T + b$$

## 4. Implementación Práctica: Una Neurona Simple (Perceptrón)

Vamos a simular el funcionamiento de una única neurona artificial que realiza una clasificación binaria sencilla (ej. decidir si aprobar un crédito basado en dos factores: **Ingresos** y **Puntaje de Crédito**).

Utilizaremos una función de activación simple (Función Escalón) para decidir si el resultado es `1` (Aprobado) o `0` (Rechazado).
$$\text{Activación}(z) = \begin{cases} 1 & \text{si } z \geq 0 \\ 0 & \text{si } z < 0 \end{cases}$$

```python
import numpy as np
import matplotlib.pyplot as plt

class SimpleNeuron:
    def __init__(self, num_features):
        # Inicializamos pesos y sesgo de manera arbitraria para el ejemplo
        self.weights = np.array([0.5, 0.8])  # Pesos para Ingresos y Puntaje de Crédito
        self.bias = -1.2                     # Sesgo (Bias)
        
    def activate(self, z):
        # Función de activación escalón (Step Function)
        return 1 if z >= 0 else 0
    
    def predict(self, x):
        # Operación lineal: z = x1*w1 + x2*w2 + b
        z = np.dot(x, self.weights) + self.bias
        # Aplicar la función de activación
        prediction = self.activate(z)
        return prediction, z

# Inicializar nuestra neurona
neuron = SimpleNeuron(num_features=2)
print("Neurona inicializada con pesos:", neuron.weights, "y sesgo:", neuron.bias)
```

```python
# Definamos algunos casos de prueba representativos: [Ingresos Escala (0-2), Score Crédito Escala (0-2)]
# Caso A: Ingresos altos, buen score
caso_a = np.array([1.5, 1.2]) 
# Caso B: Ingresos muy bajos, mal score
caso_b = np.array([0.2, 0.4]) 

pred_a, z_a = neuron.predict(caso_a)
pred_b, z_b = neuron.predict(caso_b)

print(f"Caso A (Fuerte): Valor lineal z={z_a:.2f} -> Predicción={pred_a} ({'Aprobado' if pred_a == 1 else 'Rechazado'})")
print(f"Caso B (Débil): Valor lineal z={z_b:.2f} -> Predicción={pred_b} ({'Aprobado' if pred_b == 1 else 'Rechazado'})")
```

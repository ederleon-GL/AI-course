# 01 - MCP

Base de MCP en 3 pasos.

```python
# Ejemplo mental simple de flujo MCP
flujo_mcp = {
    "1": "Descubrir herramientas disponibles",
    "2": "Leer el esquema de una herramienta",
    "3": "Preparar argumentos válidos",
    "4": "Invocar herramienta",
    "5": "Usar el resultado en la respuesta final",
}

for paso, descripcion in flujo_mcp.items():
    print(f"Paso {paso}: {descripcion}")
```

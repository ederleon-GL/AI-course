Demo rápida con tool de clima.

## Paso 1

Tool `get_weather`: `city` + `unit`.

```python
def get_weather(city, unit="C"):
    # Simulación local de respuesta MCP
    data = {
        "bogota": 18,
        "medellin": 24,
        "mexico": 21,
    }

    temp_c = data.get(city.lower())
    if temp_c is None:
        return {"ok": False, "error": "Ciudad no encontrada"}

    if unit == "F":
        temp = (temp_c * 9 / 5) + 32
    else:
        temp = temp_c

    return {
        "ok": True,
        "city": city,
        "temperature": round(temp, 1),
        "unit": unit,
    }


print(get_weather("Bogota"))
print(get_weather("Bogota", unit="F"))
```

## Paso 2

Pregunta -> detectar -> llamar tool -> responder.

```python
def responder_pregunta(pregunta):
    pregunta_low = pregunta.lower()

    if "clima" in pregunta_low or "temperatura" in pregunta_low:
        # Regla mínima para detectar ciudad
        for ciudad in ["bogota", "medellin", "mexico"]:
            if ciudad in pregunta_low:
                result = get_weather(ciudad)
                if not result["ok"]:
                    return f"No pude obtener el clima: {result['error']}"
                return (
                    f"En {result['city'].title()} hay {result['temperature']} "
                    f"grados {result['unit']}."
                )
        return "Puedo consultar clima, pero necesito que indiques una ciudad."

    return "Puedo ayudarte con preguntas de clima."


print(responder_pregunta("¿Cuál es el clima en Bogota?"))
print(responder_pregunta("hola"))
```

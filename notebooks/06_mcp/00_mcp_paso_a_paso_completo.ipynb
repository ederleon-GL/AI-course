## Paso 1: Qué es MCP

MCP conecta un agente con herramientas externas de forma estándar.

Flujo base:
1. Entender intención del usuario.
2. Seleccionar tool.
3. Validar argumentos.
4. Ejecutar tool.
5. Responder.

## Paso 2: Configuración mínima

- OpenAI: define `OPENAI_API_KEY`.
- Ollama: servicio activo en `http://localhost:11434`.

Este notebook funciona sin llamar modelo (modo local), pero incluye sección opcional con OpenAI/Ollama.

```python
# Paso 3: Tool MCP simulada (clima)
def get_weather(city, unit="C"):
    data = {
        "bogota": 18,
        "medellin": 24,
        "mexico": 21,
    }

    temp_c = data.get(city.lower())
    if temp_c is None:
        return {"ok": False, "error": "Ciudad no encontrada"}

    if unit == "F":
        temp = (temp_c * 9 / 5) + 32
    else:
        temp = temp_c

    return {
        "ok": True,
        "city": city,
        "temperature": round(temp, 1),
        "unit": unit,
    }

print(get_weather("Bogota"))
print(get_weather("Bogota", "F"))
```

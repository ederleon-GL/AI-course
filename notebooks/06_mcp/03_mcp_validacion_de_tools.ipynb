# 03 - Validación

Validar args antes de llamar.

## Paso 1

`query` obligatorio, `limit` opcional.

```python
tool_schema = {
    "required": ["query"],
    "properties": {
        "query": {"type": str},
        "limit": {"type": int, "min": 1, "max": 20},
    },
}

print("Schema cargado:")
print(tool_schema)
```

```python
def validar_argumentos(args, schema):
    # 1) Validar requeridos
    for req in schema["required"]:
        if req not in args:
            return False, f"Falta parámetro obligatorio: {req}"

    # 2) Validar tipos y rango básico
    props = schema["properties"]

    if "query" in args and not isinstance(args["query"], props["query"]["type"]):
        return False, "query debe ser string"

    if "limit" in args:
        if not isinstance(args["limit"], props["limit"]["type"]):
            return False, "limit debe ser int"
        if not (props["limit"]["min"] <= args["limit"] <= props["limit"]["max"]):
            return False, "limit fuera de rango (1-20)"

    return True, "ok"


casos = [
    {"query": "MCP"},
    {"query": "MCP", "limit": 5},
    {"query": "MCP", "limit": 100},
    {"limit": 2},
]

for caso in casos:
    valido, mensaje = validar_argumentos(caso, tool_schema)
    print(caso, "=>", valido, "-", mensaje)
```
